```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import re
from sentence_transformers import SentenceTransformer
import streamlit as st
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')

# Load data
@st.cache_data
def load_data(file_path='../100_sampled_cleaned.csv'):
    df = pd.read_csv(file_path)
    return df

df = load_data()

# Preprocessing
def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    return text.strip()

df['text_combined'] = (df['name'].apply(clean_text) + ' ' + 
                       df['description'].apply(clean_text).fillna('') + ' ' + 
                       df['specifications'].apply(clean_text).fillna(''))

# Store original price
df['price_original'] = df['price'].copy()

# Normalize numerical
scaler = StandardScaler()
num_features = ['price', 'rating_avg']
df[num_features] = scaler.fit_transform(df[num_features].fillna(0))

# Taxonomy
def parse_taxonomy(breadcrumbs):
    if pd.isna(breadcrumbs):
        return ['']
    return breadcrumbs.split(' > ')

df['taxonomy_levels'] = df['breadcrumbs'].apply(parse_taxonomy)
df['top_category'] = df['taxonomy_levels'].apply(lambda x: x[0] if x else '')

# Feature Engineering
@st.cache_data
def get_features():
    tfidf = TfidfVectorizer(max_features=1000)
    text_vectors = tfidf.fit_transform(df['text_combined'])
    
    model_sbert = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
    text_embeddings = model_sbert.encode(df['text_combined'].tolist())
    
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    df['category_encoded'] = le.fit_transform(df['top_category'])
    
    combined_features = np.hstack([text_vectors.toarray(), df[num_features], df['category_encoded'].values.reshape(-1,1)])
    return combined_features, text_embeddings, tfidf

combined_features, text_embeddings, tfidf = get_features()

# Similarity
similarity_matrix = cosine_similarity(combined_features)

# Functions
def find_matching_product(input_name, threshold=0.1):
    input_clean = clean_text(input_name)
    input_vec = tfidf.transform([input_clean])
    sim_scores = cosine_similarity(input_vec, combined_features)[0]
    best_idx = np.argmax(sim_scores)
    if sim_scores[best_idx] > threshold:
        return df.iloc[best_idx]['product_id'], sim_scores[best_idx]
    return None, 0

def get_recs_from_list(viewed_ids, k=10, sim_matrix=similarity_matrix):
    viewed_indices = [df[df['product_id'] == pid].index[0] for pid in viewed_ids if pid in df['product_id'].values]
    if not viewed_indices:
        return []
    # Average similarity from viewed
    avg_sim = np.mean([sim_matrix[i] for i in viewed_indices], axis=0)
    # Exclude viewed
    mask = np.ones(len(df), dtype=bool)
    mask[viewed_indices] = False
    avg_sim[~mask] = -1  # Penalize non-masked
    top_indices = np.argsort(avg_sim)[::-1][:k]
    rec_ids = df.iloc[top_indices]['product_id'].tolist()
    return rec_ids

def get_recs(product_id, k=10, sim_matrix=similarity_matrix):
    idx = df[df['product_id'] == product_id].index[0]
    sim_scores = list(enumerate(sim_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:k+1]
    rec_ids = [df.iloc[i[0]]['product_id'] for i in sim_scores]
    return rec_ids

# Streamlit App
st.title("Product Recommendation Demo")

tab1, tab2 = st.tabs(["Input by Name", "Input by Viewed List"])

with tab1:
    input_name = st.text_input("Enter Product Name (e.g., 'Ví Nam')")
    if input_name:
        match_id, match_score = find_matching_product(input_name)
        if match_id:
            st.success(f"Matching Product: ID {match_id} (Similarity: {match_score:.2f})")
            rec_ids = get_recs(match_id, 10)
            
            st.subheader("Top 10 Recommendations")
            rec_df = df[df['product_id'].isin(rec_ids)][['product_id', 'name', 'price_original', 'top_category']]
            st.dataframe(rec_df)
        else:
            st.warning("No matching product found. Try a different name.")

with tab2:
    viewed_input = st.text_area("Enter Viewed Product IDs (comma-separated, e.g., 2000337,191389544)")
    if viewed_input:
        try:
            viewed_ids = [int(x.strip()) for x in viewed_input.split(',')]
            rec_ids = get_recs_from_list(viewed_ids, 10)
            if rec_ids:
                st.subheader("Top 10 Recommendations")
                rec_df = df[df['product_id'].isin(rec_ids)][['product_id', 'name', 'price_original', 'top_category']]
                st.dataframe(rec_df)
            else:
                st.warning("No recommendations or invalid IDs.")
        except ValueError:
            st.error("Invalid input. Please enter integers separated by commas.")

# Common filters and charts
st.sidebar.header("Filters")
price_max = float(df['price_original'].max())
price_filter = st.sidebar.slider("Max Price", 0.0, price_max, price_max, step=1.0)
cat_filter = st.sidebar.multiselect("Categories", df['top_category'].unique())
filtered_df = df[df['price_original'] <= price_filter]
if cat_filter:
    filtered_df = filtered_df[filtered_df['top_category'].isin(cat_filter)]

st.subheader("Filtered Products")
st.dataframe(filtered_df[['product_id', 'name', 'price_original']])

col1, col2 = st.columns(2)
with col1:
    fig1 = px.histogram(df, x='price_original', title="Price Distribution")
    st.plotly_chart(fig1)
with col2:
    fig2 = px.pie(df, names='top_category', title="Category Coverage")
    st.plotly_chart(fig2)
```
